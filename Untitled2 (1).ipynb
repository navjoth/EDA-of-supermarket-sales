# Exploratory Data Analysis Of Supermarket Sales

### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Reading the CSV dataset

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv')
df.shape
```

```python
df
```

#### Data cleaning

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['Time'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.rename(columns={'Product line':'Category','cogs':'COGS'},inplace=True)
df
```

```python
df[['Unit price','Quantity','Total','gross income','Rating']].describe()
```

### Exploratory Data Analysis

#### Gender

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
total_gen = df.groupby(['Gender'],as_index=False)['Total'].sum().sort_values(by=['Total'],ascending=False)
total_gen
```

```python
sns.barplot(x='Gender',y='Total', hue='Gender',data=total_gen, palette=['pink','blue'])
```

#### Plotting Total Sales by Cities

```python
city_sales = df.groupby('City')['Total'].sum()
plt.figure(figsize=(5, 5))
plt.pie(city_sales, labels=city_sales.index, autopct='%1.1f%%', colors=['red', 'blue', 'green'], startangle=90)
plt.title('Total Sales Distribution Across Cities')
plt.show()
```

```python
sns.barplot(x='City',y='Total', hue='City',data=df, palette=['red','green','blue'])
plt.title('Distribution of Total Sales across Cities')
plt.show()
```

#### Distribution Of Categories

```python
df['Category'].unique()
```

```python
category_sales = df.groupby(['Category', 'Branch'])['Quantity'].sum().reset_index()
sns.barplot(x='Category', y='Quantity', hue='Branch', data=category_sales)
plt.title('Total Quantity by Category and Branch')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

#### Time Series Trend

```python
df.set_index('Date').resample('ME')['Total'].sum().plot()
plt.title('Monthly Sales Trend')
plt.show()
```

#### Members vs Non Members sales

```python
sns.boxplot(x='Customer type', y='Total', data=df)
plt.title('Customer Type vs Total Sales')
plt.show()
```

#### Payment Mode Analysis

```python
payment_mode_sales = df.groupby('Payment')['Total'].mean().sort_values(ascending=False)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
payments=sns.barplot(x=payment_mode_sales.index, y=payment_mode_sales.values,hue=payment_mode_sales.index, palette=colors)
for bars in payments.containers:
    payments.bar_label(bars)
plt.title('Average Sales by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

#### Weekday vs Weekend sales analysis

```python
df['day_of_week'] = df['Date'].dt.day_name()
weekday_sales = df.groupby('day_of_week')['Total'].sum().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
sns.barplot(x=weekday_sales.index, y=weekday_sales.values)
plt.title('Sales by Day of the Week')
plt.show()
```

#### Rating

```python
sns.scatterplot(x='Total', y='Rating', data=df)
plt.title('Total Sales vs Rating')
plt.show()
```

```python
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
rt=sns.barplot(x=category_ratings.values, y=category_ratings.index)
for bars in rt.containers:
    rt.bar_label(bars)
plt.title('Average Rating by Category')
plt.xticks(rotation=45)
plt.show()
```
